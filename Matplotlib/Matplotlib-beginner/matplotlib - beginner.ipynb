```python
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='solid', linewidth=6)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted', linewidth=2)
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='solid', linewidth=2, marker='o')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted', linewidth=2, marker='.')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='solid', linewidth=2, marker='o', markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted', linewidth=2, marker='.', markersize=10)
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='solid', linewidth=2, marker='o', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted', linewidth=2, marker='.', markersize=10, label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend()
plt.show()
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#f7ce00', linestyle='solid', linewidth=2, marker='o', markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#fa05d5', linestyle='dotted', linewidth=2, marker='.', markersize=10, label='Rohit')
plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
plt.legend(loc='lower center')
plt.show()
```

## Bar Chart

 - Bivariate Analysis
 - numerical vs categorical
 - use case - Aggregate Analysis of Groups

```python
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
```

```python
plt.bar(colors, children)
plt.show()
```

```python
plt.bar(colors, children, color='black', align='center')
plt.title('Colors VS Number of Children')
plt.xlabel('Colors')
plt.ylabel('Number of Children')
plt.grid()
plt.show()
```

```python
plt.bar(colors, children, color='black', align='edge', bottom=10)
plt.title('Colors VS Number of Children')
plt.xlabel('Colors')
plt.ylabel('Number of Children')
plt.grid()
plt.show()
```

```python
plt.barh(colors, children, color='black', align='center')
plt.title('Colors VS Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Colors')
plt.grid()
plt.show()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])
plt.show()
```

```python
plt.bar(df['batsman'], df['2015'], width=0.1)
plt.show()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.5, df['2015'], width=0.1)
plt.bar(np.arange(df.shape[0]) - 0.3, df['2016'], width=0.1)
plt.bar(np.arange(df.shape[0]) - 0.1, df['2017'], width=0.1)
plt.show()
```

```python
np.arange(df.shape[0])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='#f507b2', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='#07e5f5', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='#f1f507', label='2017')
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.title('Batsman VS Runs Score')
plt.xlabel('Batsman')
plt.ylabel('Runs Score')
plt.legend()
plt.grid()
plt.show()
```

```python
children = [10, 20, 40, 10, 30]
colors = ['red red red red red', 'blue blue blue', 'green green green', 'yellow yellow', 'pink pink pink pink']

plt.bar(colors, children, color='black')
plt.xticks(rotation='vertical')
plt.show()
```

## Pie Chart

 - Univaritate/Bivariate Analysis
 - categorical vs numerical
 - use case - To find contribution on a standard scale

```python
data = [23, 45, 100, 20, 49]

plt.pie(data)
plt.show()
```

```python
data = [23, 45, 100, 20, 49]
subjects = ['eng', 'science', 'maths', 'sst', 'hindi']
plt.pie(data, labels=subjects, autopct='%0.1f%%', colors=['#a7511b', '#fe4728', '#ff8a82', '#faa41b', '#ffd45b'], shadow=True, explode=[0.3, 0, 0, 0, 0])
plt.title("Marks of Subjects")
plt.show()
```
