```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
species_map = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
iris['Species'] = iris['Species'].map(species_map)
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], 
            iris['PetalWidthCm'], 
            c=iris['Species'],
            cmap='viridis')
plt.title('Sepal Length VS Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], 
            iris['PetalWidthCm'], 
            c=iris['Species'], 
            cmap='cool')
plt.title('Sepal Length VS Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], 
            iris['PetalWidthCm'], 
            c=iris['Species'], 
            cmap='cool')
plt.title('Sepal Length VS Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.colorbar()
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], 
            iris['PetalWidthCm'], 
            c=iris['Species'], 
            cmap='cool', 
            alpha=0.6)
plt.title('Sepal Length VS Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.colorbar()
plt.show()
```

## 3D line Plot

```python
x = [0, 1, 5, 4]
y = [0, 10, 13, 2]
z = [0, 13, 20, 0]

fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x, y, z, color='blue', s=[100, 100, 100, 100])
ax.plot3D(x, y, z, color='red')
ax.set_title('3D Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z', rotation=90, labelpad=-2)
plt.show()
```

## 3D Surface Plots

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
```

```python
z.shape
```

```python
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='rocket')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='rocket')
fig.colorbar(p)
plt.show()
```

## Heatmap

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin(range(1, 7))) & (delivery['batsman_run']==6)]
```

```python
temp_df
```

```python
temp_df['ballnumber'].unique()
```

```python
grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
```

```python
grid
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(grid, cmap='rocket')
plt.colorbar()
plt.show()
```

```python
plt.figure()
plt.xticks(np.arange(0, 6), list(range(1, 7)))
plt.yticks(delivery['overs'].unique(), list(range(1, 21)))
plt.imshow(grid, cmap='magma')
plt.colorbar()
plt.show()
```
