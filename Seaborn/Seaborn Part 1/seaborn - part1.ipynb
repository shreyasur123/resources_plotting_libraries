## Relational Plots

 - to see the statistical realtion between 2 or more variables
 - Bivariate Analysis
#### Plots under this section
 - scatterplot
 - lineplot

```python
# figure level function -> relplot
# axis level functions -> scatterplot, lineplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# scatterplot -> axis level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# relplot -> figure level function(recommended by seaborn) -> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
plt.show()
```

## Distribution Plots

 - used for univariate analysis
 - used to find out the distribution
 - Range of the observation
 - Central Tendency
 - is the data bimodal?
 - Are there outliers?
#### Plots under distribution plot
 - histplot
 - kdeplot
 - rugplot

```python
# figure level function -> displot
# axis level function -> histplot, kdeplot, rugplot
```

```python
tips
```

```python
# histplot
sns.histplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='total_bill', bins=20)
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='total_bill', bins=2)
plt.show()
```

```python
tips
```

```python
# countplot
sns.displot(kind='hist', data=tips, x='day')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='tip', hue='sex', element='step')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(kind='hist', data=titanic, x='age', element='step', hue='sex')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='smoker', element='step', col='sex')
plt.show()
```

```python
sns.displot(kind='hist', data=tips, x='tip', element='step', col='sex')
plt.show()
```

```python
# kdeplot
sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=10)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=10, aspect=2)
plt.show()
```

```python
# rugplot
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
plt.show()
```

```python
# bivariate histogram
sns.histplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, kind='hist', x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, kind='kde', x='total_bill', y='tip')
plt.show()
```

## Matrix Plots

 - Heatmap
 - Clustermap

```python
# heatmap -> axis level function
# there's no figure level function for heatmap
gap
```

```python
temp_df = gap.pivot(index='country', columns='year', values='lifeExp')
```

```python
temp_df
```

```python
sns.heatmap(temp_df)
plt.show()
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(temp_df)
plt.show()
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
```

```python
temp_df
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(temp_df, annot=True,linewidth=0.5, cmap='summer')
plt.show()
```

```python
# clustermap

iris = px.data.iris()
```

```python
iris
```

```python
sns.clustermap(iris.iloc[:, :-2])
plt.show()
```
